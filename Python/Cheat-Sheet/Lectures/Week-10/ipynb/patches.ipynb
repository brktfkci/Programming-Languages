```python
#personalize a path
import matplotlib.patches as patches
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
path = [
    [.1, .3],
    [.2, .9],
    [.8, .4],
]
ax.add_patch(patches.Polygon(
  path,
  alpha=.7,
  edgecolor="darkblue",
  facecolor="red",
  hatch="-",
  joinstyle="round",
  linestyle="--",
  linewidth=5,
  fill=False,
  visible=True
)) 
plt.show()
```

```python
#more polygons
#--------------
import numpy as np
import matplotlib.patches as patches
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

theta = np.linspace(0, 2 * np.pi, 8)
points = np.vstack((np.cos(theta), np.sin(theta))).transpose()
ax.add_patch(patches.Polygon(points, color = 'yellow'))
ax.grid(True)
ax.axis('scaled')
plt.show()
```

```python
#adding geometric objects to plot area
#the following function draws a line between two points
#mlines can be used instead of patches
#---------------------
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

def newline(p1, p2):
    ax = plt.gca() #get current axes
#get_xbound is used to retrieve x limits
    xmin, xmax = ax.get_xbound() 

    if(p2[0] == p1[0]):
        xmin = xmax = p1[0]
#get_ybound is used to retrieve y limits
        ymin, ymax = ax.get_ybound()
    else:
        ymax = p1[1]+(p2[1]-p1[1])/(p2[0]-p1[0])*(xmax-p1[0])
        ymin = p1[1]+(p2[1]-p1[1])/(p2[0]-p1[0])*(xmin-p1[0])

    l = mlines.Line2D([xmin,xmax], [ymin,ymax])
    ax.add_line(l)
    return l
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.patches as patches

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

x = np.linspace(0,10); y = x**2
p1 = [2,10]; p2 = [7,70]
ll=mlines.Line2D(p1, p2, color="r", ls = '-.', lw = 3)
ax.text(3,50, "Hello OzU", fontsize=18, fontfamily="Times", color="b")
ax.plot(x, y)
ax.add_line(ll)

rect = patches.Rectangle((2,70),5,20,linewidth=1,edgecolor='r',facecolor='yellow')
circ = patches.Circle((0,25), 5, fc='y')
ax.add_patch(rect)
ax.add_patch(circ)
plt.show()
```

```python
#-----------------
#more shapes
#-----------------
import matplotlib.patches as patches
import matplotlib.pyplot as plt

# Circle
shape = patches.Circle((0, 0), radius = 1., color = '.75')
plt.gca().add_patch(shape)

# Rectangle
shape = patches.Rectangle((2.5, -.5), 2., 1., color = '.75')
plt.gca().add_patch(shape)

# Ellipse
shape = patches.Ellipse((0, -2.), 2., 1., 
                        angle = 45., color = '.75')
plt.gca().add_patch(shape)

# Fancy box
shape = patches.FancyBboxPatch(
    (2.5, -2.5), 2., 1., 
    boxstyle = 'sawtooth', color = '.75')
plt.gca().add_patch(shape)

# Display all
plt.grid(True)
plt.axis('scaled')
plt.show()
```

```python
#------------
#example from https://matplotlib.org/
#------------
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(4,4), dpi=150)

Path = mpath.Path
path_data = [
    (Path.MOVETO, (1.58, -2.57)),
    (Path.CURVE4, (0.35, -1.1)),
    (Path.CURVE4, (-1.75, 2.0)),
    (Path.CURVE4, (0.375, 2.0)),
    (Path.LINETO, (0.85, 1.15)),
    (Path.CURVE4, (2.2, 3.2)),
    (Path.CURVE4, (3, 0.05)),
    (Path.CURVE4, (2.0, -0.5)),
    (Path.CLOSEPOLY, (1.58, -2.57)),
    ]
codes, verts = zip(*path_data)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path, facecolor='r', alpha=0.5)
ax.add_patch(patch)

# plot control points and connecting lines
x, y = zip(*path.vertices)
line, = ax.plot(x, y, 'go-')

ax.grid()
ax.axis('equal')
plt.show()
```
