```python
#------working with major and minor tickmarks
#EE393
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
fig = plt.figure(figsize=(2.5,2.5), dpi=150)
ax = fig.add_axes([0,0,1,1])

X = np.linspace(-15, 15, 500)
Y = np.sinc(X)        #sinc(x)=sin(pi*x)/pi*x
ax.set_xlim(-15,15); ax.set_ylim(-0.25,1.0)
ax.set_xlabel('X'); ax.set_ylabel('sinc(X)')
labelsX=np.arange(-15,15,3)

#major tickmarks at every 5 
#minor tickmarks at every 2.5 
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(2.5))

ax.yaxis.set_major_locator(ticker.MultipleLocator(0.25))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.05))

#ax.set_xticklabels(labelsX,rotation=45)
ax.tick_params(axis='x', labelrotation=45)
ax.tick_params(axis='y', labelrotation=45, color="red")

plt.plot(X, Y, color = 'green')
plt.show()
```

```python
#twin axes
#below, two graphs share one x axis; but y axes are different
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure(figsize=(3,3), dpi=150, 
                 facecolor="lightgrey", edgecolor="red", linewidth=5) 
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='center')
plt.show()
```

```python
#managing sub-plots
#(partly taken from tutorialpoints)
import matplotlib.pyplot as plt
import numpy as np
'''
+------+-----+
| (0,0)|(0,1)|
|------+-----+
| (1,0)|(1,1)|
+------------+
'''
fig,ff =  plt.subplots(2, 2,   #2 rows, 2 columns
                 figsize=(5,5), dpi=150, 
                 facecolor="lightgrey", 
                 edgecolor="red", linewidth=5) 
#ff holds axes references for subplots

x = np.arange(1,5) #x values
ff[0][0].plot(x,x*x) #plot at (0,0)
ff[0][0].set_title('square')
ff[0][1].plot(x,np.sqrt(x)) #plot at (0,1)
ff[0][1].set_title('square root')
ff[1][0].plot(x,np.exp(x)) #plot at (1,0)
ff[1][0].set_title('exp')
ff[1][1].plot(x,np.log10(x)) #plot at (1,1)
ff[1][1].set_title('log')
plt.show()
```

```python
#arranging grids in multi plots
#(taken from tutorialpoints)
import matplotlib.pyplot as plt
import numpy as np
'''
+------+-----+------+
| (0)  | (1) | (2)  |
|------+-----+------+
'''
fig, axes = plt.subplots(1,3, figsize = (12,4))
x = np.arange(1,11)
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls = '-.', lw = 0.25)
axes[1].set_title('custom grid')
axes[2].plot(x,x)
axes[2].set_title('no grid')
fig.tight_layout()
plt.show()
```

```python
# a 3X3 grid of the figure object is filled with 
# axes objects of varying sizes in row and column 
# spans, each showing a different plot.
import matplotlib.pyplot as plt
'''
+------+-----+------+
| first      |      |
|------+-----+second+
|            |      |
+ third      |      |  
|            |      |
+------+-----+------+
first: starts at (0,0), spans two columns; 
second: starts at (0,2), spans three rows
third: starts at (1,0), spans two columns and two rows;
'''
fig = plt.figure(figsize=(4,4), dpi=150, 
                 facecolor="floralwhite", edgecolor="white", linewidth=5)
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2)
import numpy as np
x = np.arange(1,10)
a1.plot(x, np.exp(x))
a1.set_title('exp')
a2.plot(x, x*x)
a2.set_title('square')
a3.plot(x, np.log(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```

```python
#an advanced example for multi-figures
#adapted from paxt website
import numpy as np
from matplotlib import pyplot as plt
def get_radius(T, params):  
    m, n_1, n_2, n_3 = params  
    U = (m * T) / 4  
    return (np.fabs(np.cos(U)) ** n_2 + 
            np.fabs(np.sin(U)) ** n_3) ** (-1. / n_1)

sizeX,sizeY = 3, 4
fig,ax = plt.subplots(sizeX,sizeY,figsize=(5,5), dpi=150) 
T = np.linspace(0, 2 * np.pi, 1024)
for i in range(sizeX):  
    for j in range(sizeY):    
        params = np.random.randint(1, 20, size = 4)    
        R = get_radius(T, params)      
        ax[i][j].get_xaxis().set_visible(False)    
        ax[i][j].get_yaxis().set_visible(False)    
        ax[i][j].plot(R * np.cos(T), R * np.sin(T), color = 'black')   
        ax[i][j].set_title(tuple(params), fontsize = 'small')
plt.tight_layout()
plt.show()
```

```python
#Adding arrows to your plots
#plot anootations
#-------------
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(3,3), dpi=150) 
ax = fig.add_axes([0,0,1,1]) #axes
X = np.linspace(-4, 4, 1000)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)
#xytext: text coordinate; xy:where the arrow points to
#shrink: shrink factor.
ax.annotate('Minimum value',
             xytext = (-1.5, 3.),
             xy = (0.75, -2.7),
             arrowprops = { 'facecolor' : 'red', 
                           'edgecolor': 'red',
                           'arrowstyle': '->'
                          })
ax.plot(X, Y)
plt.show()
```

```python
#-----------
#advanced
#3D extension to matplotlib
#-----------------
import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(3,3), dpi=150) 
ax = fig.add_axes([0,0,1,1], projection = '3d') #axes
x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
X, Y = np.meshgrid(x, y)
Z = np.sinc(np.sqrt(X ** 2 + Y ** 2))

ax.plot_surface(X, Y, Z, cmap=cm.Greens)
plt.show()
```
